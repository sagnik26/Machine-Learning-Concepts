```python
import pandas as pd

df = pd.read_csv("data.csv")

df
```

# Selecting Data


```python
print("First 5 rows \n", df.head(), "\n")        # first 5 rows
print("(rows, columns) \n", df.shape, "\n")        # (rows, columns)
print("df.shape \n", df.dtypes, "\n")       # column types
print("stats \n", df.describe(), "\n")     # stats: mean, std, min, max...
print("count missing values per column \n", df.isnull().sum(), "\n") # count missing values per column
```

```python
print("age column", df[["age"]])

print("filter rows where age > 30", df[df["age"] > 30])
```

```python
print("loc: rows by condition, specific columns", df.loc[df["hired"] == True, ["age", "salary"]])
```

# Cleaning Data

```python
print("drop rows with any missing values ", df.dropna())

print("fill missing values with the column mean \n", df["salary"].fillna(df["salary"].mean()))

print("fill in-place NAN values across the whole df \n", df.fillna(0))

print("rename a column \n", df.rename(columns={"hired": "label"}))

print("drop a column \n", df.drop(columns=["age"]))
```
